```python
import pandas as pd
import numpy as py
import seaborn as sns
import os
```

```python
safety_data_df = pd.read_csv('safety.csv')
safety_data_df.head()
```

```python
#check for null values or empty cells
null_values = safety_data_df.isnull().any()
```

```python
#drop null values or empty cells
safety_data_df.dropna()
```

```python
safety_data_df.count()
```

```python
#Save the cleaned dataset to a new csv file
safety_data_df.to_csv('Safety_cleaned_data.csv', index=False)
```

```python
#sorting by timeperiod
safety_data_df = safety_data_df.sort_values(['2022'])
safety_data_df
```

```python
safety_data_df = safety_data_df.sort_values(['2010'])
safety_data_df
```

```python
safety_data_df["change"] = safety_data_df['2022']-safety_data_df['2010']
safety_data_df["change"]
```

```python
# calculate growth rate for 10 years
safety_data_df['growth_rate'] = safety_data_df['change'].pct_change() * 100
safety_data_df['growth_rate']
```
